Defaultdict 
In which we can create the default dictionaries and can added the key and the value in it independantly 


```python
from  collections import defaultdict
my_list=["apple","orange","pineapple","banana","kiwi","mango","avocado"]
my_dict=defaultdict(list)
for item in my_list:
    my_dict[item[0]].append(item)
print(my_dict)
```

##Counters:- 
Control turns a sequence of values into a defaultdict(int)-like object mapping keys to counts
same as the default dict as well . just check the frequency of the that element in the list in terms of the dictionaries

```python
from collections import Counter
c=Counter([0,1,2,3,2,0])
print(c[2])
print(c[1])
```

Assert  used for the testing and debugging

```python
##assert condition, "Optional error message"
import numpy as np

X = np.array([[1, 2], [3, 4]])

# Verify input shape for a model
assert X.shape[1] == 2, "Input must have exactly 2 features"
assert X[0][1]==2, "Answer should be the 2"
```
