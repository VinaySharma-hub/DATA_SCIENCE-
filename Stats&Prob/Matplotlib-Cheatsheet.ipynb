# Making a Random Dataset for Visualization

```python
import pandas as pd
import numpy as np
# Creating the data dictionary
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [23, 17, 35, 29],
    'Year': [2020, 2021, 2022, 2023],
    'Percentage': [50, 30, 60, 40],
    'Group': ['X', 'Y', 'X', 'Y']
}

# Creating the DataFrame
beginner_df = pd.DataFrame(data)

# Display the DataFrame
beginner_df.head()
```

# Beginners Level

## 1 Line Plot

```python
plt.figure(figsize=(8, 5))
plt.plot(beginner_df['Year'], beginner_df['Values'], marker='o')
plt.title('Values Over Years')
plt.xlabel('Year')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

## 2 Bar Plot

```python
plt.figure(figsize=(8, 5))
plt.bar(beginner_df['Category'], beginner_df['Values'], color='skyblue')
plt.title('Values by Category')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

## 3 Scatter Plot

```python
plt.figure(figsize=(8, 5))
plt.scatter(beginner_df['Year'], beginner_df['Percentage'], color='red')
plt.title('Percentage Over Years')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.show()
```

## 4 Pie Chart

```python
plt.figure(figsize=(8, 5))
plt.pie(beginner_df['Values'], labels=beginner_df['Category'], autopct='%1.1f%%')
plt.title('Values Distribution by Category')
plt.show()
```

## 5 Histogram

```python
plt.figure(figsize=(8, 5))
plt.hist(beginner_df['Values'], bins=5, color='purple', edgecolor='black')
plt.title('Values Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

## 6 Box Plot
Useful for visualizing the distribution and detecting outliers.

## 12 Stacked Bar Plot
Visualize the composition of different groups in a category.

# Intermediate Level

## 1 Subplots with Shared Axes
Creating subplots that share the same x or y-axis to compare data more easily.

```python
# Creating sample data for subplots
years = [2018, 2019, 2020, 2021, 2022]
group_x = [12, 15, 18, 20, 25]
group_y = [8, 10, 15, 17, 22]

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(8, 6))

# First subplot
ax1.plot(years, group_x, marker='o', color='blue')
ax1.set_title('Group X over Years')
ax1.grid(True)

# Second subplot
ax2.plot(years, group_y, marker='o', color='green')
ax2.set_title('Group Y over Years')
ax2.grid(True)
ax2.set_xlabel('Year')

plt.tight_layout()
plt.show()
```

## 2 Twin Axes Plot
Creating two y-axes on the same plot to show data with different scales.

```python
# Sample data
x = [2020, 2021, 2022, 2023]
y1 = [30, 35, 40, 45]  # First dataset
y2 = [100, 110, 105, 115]  # Second dataset with different scale

fig, ax1 = plt.subplots(figsize=(8, 5))

# Plotting first dataset
ax1.plot(x, y1, color='blue', marker='o')
ax1.set_ylabel('Values Y1 (Blue)', color='blue')
ax1.set_xlabel('Year')
ax1.grid(True)

# Adding second y-axis
ax2 = ax1.twinx()
ax2.plot(x, y2, color='red', marker='o', linestyle='--')
ax2.set_ylabel('Values Y2 (Red)', color='red')

plt.title('Twin Axes Plot')
plt.show()
```

## 5 Waterfall Chart
A variation of the bar chart to show how an initial value is affected by intermediate positive or negative values.

```python
# Sample data for waterfall chart
categories = ['Start', 'Product A', 'Product B', 'Discount', 'End']
values = [100, 20, 15, -10, 125]

# Calculating cumulative values
cumulative = np.cumsum(values)

# Colors for different bars
colors = ['skyblue' if v >= 0 else 'red' for v in values]

plt.figure(figsize=(8, 5))
plt.bar(categories, cumulative, color=colors)
plt.plot(cumulative, marker='o', linestyle='--', color='grey')
plt.title('Waterfall Chart')
plt.xlabel('Category')
plt.ylabel('Cumulative Values')
plt.grid(True)
plt.show()
```

## 6 Radar Chart (Spider Plot)
Visualize multivariate data in a circular format.

```python
# Sample data for radar chart
labels = ['A', 'B', 'C', 'D', 'E']
values = [4, 3, 2, 5, 4]

# Adding a repeated value to close the circle
values += values[:1]

# Calculating angles for each axis
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]

# Plotting
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'polar': True})
ax.fill(angles, values, color='skyblue', alpha=0.4)
ax.plot(angles, values, color='blue', marker='o')
ax.set_yticks([1, 2, 3, 4, 5])
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
plt.title('Radar Chart')
plt.show()
```

## 24 Streamplot
Visualizing the flow of data, such as wind or water currents.

```python
# Sample data for streamplot
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2

plt.figure(figsize=(8, 6))
plt.streamplot(X, Y, U, V, color=np.sqrt(U**2 + V**2), cmap='autumn')
plt.title('Streamplot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## 1 Interactive Plotting with Widgets
Using ipympl and ipywidgets to create interactive plots within Jupyter notebooks.

```python
# First, install ipympl if not already installed: pip install ipympl ipywidgets

import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Generating sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Interactive function
def update_plot(freq=1.0, amplitude=1.0):
    plt.figure(figsize=(8, 4))
    plt.plot(x, amplitude * np.sin(freq * x), color='blue')
    plt.title(f'Sine Wave with Frequency = {freq} and Amplitude = {amplitude}')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

# Creating interactive widgets
interact(update_plot, freq=(0.1, 5.0, 0.1), amplitude=(0.1, 2.0, 0.1))
```

```python
# Sample data for radar chart
categories = ['Speed', 'Reliability', 'Comfort', 'Safety', 'Efficiency']
values = [4, 3, 2, 5, 4]
values += values[:1]

# Creating a radar chart
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='teal', alpha=0.25)
ax.plot(angles, values, color='teal', linewidth=2)
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
plt.title('Advanced Radar Chart')
plt.show()
```

## 4 Inset Plots
Adding a smaller plot within a larger one for more detailed analysis of a specific region.

```python
# Sample data for inset plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
noise = 0.1 * np.random.randn(100)
y2 = y + noise

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, label='Original Sine Wave')
ax.plot(x, y2, label='Noisy Sine Wave')

# Adding an inset plot
ax_inset = fig.add_axes([0.5, 0.5, 0.35, 0.35])
ax_inset.plot(x, y, label='Original Sine Wave', color='blue')
ax_inset.plot(x, y2, label='Noisy Sine Wave', color='orange')
ax_inset.set_xlim(2, 4)
ax_inset.set_ylim(-1, 1)
ax_inset.set_title('Inset Plot')
ax.legend()
plt.title('Advanced Plot with Inset')
plt.show()
```

## 5 Complex Subplots and Shared Axes
Creating multiple plots with shared axes and complex grid layouts.

```python
# Sample data for complex subplots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x / 10)

fig, axs = plt.subplots(2, 2, figsize=(10, 8), sharex=True)

# Plotting in each subplot
axs[0, 0].plot(x, y1, 'r-')
axs[0, 0].set_title('Sine Function')

axs[0, 1].plot(x, y2, 'g-')
axs[0, 1].set_title('Cosine Function')

axs[1, 0].plot(x, y3, 'b-')
axs[1, 0].set_title('Tangent Function')
axs[1, 0].set_ylim(-5, 5)

axs[1, 1].plot(x, y4, 'm-')
axs[1, 1].set_title('Exponential Function')

plt.suptitle('Advanced Subplots with Shared Axes')
plt.show()
```

## 8 Quiver Plot
Visualizing vector fields using arrows to represent the direction and magnitude of data points.

```python
# Generating sample data for quiver plot
x = np.arange(-5, 5, 1)
y = np.arange(-5, 5, 1)
X, Y = np.meshgrid(x, y)
U = -1 - X**2 + Y
V = 1 + X - Y**2

plt.figure(figsize=(8, 6))
plt.quiver(X, Y, U, V, color='blue', pivot='mid')
plt.title('Advanced Quiver Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

## 9 Animated Plots
Creating animations using FuncAnimation to visualize data over time.

```python
import matplotlib.animation as animation

# Sample data for animated plot
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots()
line, = ax.plot(x, y, color='blue')
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1, 1)
ax.set_title('Animated Sine Wave')

# Update function for the animation
def update(frame):
    line.set_ydata(np.sin(x + frame / 10))
    return line,

# Create animation
ani = animation.FuncAnimation(fig, update, frames=100, interval=50, blit=True)
plt.show()
```
