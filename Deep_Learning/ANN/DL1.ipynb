```python
def perceptron(x,y):
    x=np.insert(x,0,1, axis=1)
    weights=np.ones(x.shape[1])
    lr=0.1
    for i in range(1000):
        j=np.random.randint(0,100)
        y_hat=step(np.dot(x[j],weights))
        weights=weights+lr*(y[j]-y_hat)*x[j]
    return weights[0],weights[1:]
```

```python
def step(z):
    return 1 if z>0 else 0 
```

```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
from sklearn.pipeline import Pipeline
from scikeras.wrappers import KerasClassifier

from sklearn.impute import SimpleImputer
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
data=pd.read_csv(r"C:\Users\Welcome\OneDrive\Dokumen\dsml\Deep_Learning\datasets\Churn_Modelling.csv")
data
```

```python
data=data.drop(columns=['RowNumber','CustomerId','Surname'],axis=1)
label_encoder_gender=LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
one_hot_encoder=OneHotEncoder(handle_unknown='ignore',drop='first')
geo_encoded = one_hot_encoder.fit_transform(data[['Geography']]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded,columns=one_hot_encoder.get_feature_names_out(['Geography']))
```

```python
data=pd.concat([data.drop(columns=['Geography']),geo_encoded_df])
```

```python

x=data.drop('Exited',axis=1).values
y=data['Exited'].values
np.isnan(x).sum()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)

```

```python
# X_train=X_train.dropna()
# Y_train=Y_train[X_train.index]
```

```python
# Impute missing values (if any)
imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)

# Ensure target is 1D numeric
Y_train = np.asarray(Y_train).ravel().astype(int)
```

```python
## saving encoderes for later use 
import pickle
with open('label_encoder_gender.pkl','wb')as file :
    pickle.dump(label_encoder_gender,file)
with open('OneHotEncoder.pkl','wb') as file:
    pickle.dump(OneHotEncoder,file)
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
## defie a function to create the model and try different parameter ( keras classifier)
def create_model(neurons=32,layers=1):
    model=Sequential()
    model.add(Dense(neurons,activation="relu",input_shape=(X_train.shape[1],)))
    for _ in range(layers-1):
        model.add(Dense(neurons,activation="relu"))
    model.add(Dense(1,activation='sigmoid'))
    model.compile(optimizer='adam',loss="binary_crossentropy",metrics=['accuracy'])
    return model
```

```python
## create a keras classifier 
model=KerasClassifier(build_fn=create_model,verbose=0)
```

```python
param_grid={
    'model__neurons':[16,32,64,128],
    'model__layers':[1,2],
    'epochs':[5,10]
}
```

```python
grid=GridSearchCV(estimator=model,param_grid=param_grid,n_jobs=-1,cv=3)
grid_result=grid.fit(X_train,Y_train)
```

```python
print("Best Score:", grid_result.best_score_)
print("Best Params:", grid_result.best_params_)
```
