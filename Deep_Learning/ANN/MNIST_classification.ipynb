```python
import tensorflow 
from tensorflow import keras 
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train[0]
```

```python
model = Sequential()
```

```python
model.add(Flatten(input_shape=(28,28)))
```

```python
model.add(Dense(128,activation='relu'))
```

```python
model.add(Dense(10,activation='softmax')) ## we use softmax if we need more than 2 nodes in the neural network 
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam')  ## in sparse_categorical_crossentropy we don't need to do one hot encoding
# while in categorical_crossentropy we need to do one hot encoding for sure 
```

A validation split in model fitting divides the dataset into training and validation subsets, where the model trains on the training data and evaluates performance on the unseen validation data to monitor fit quality.

```python
model.fit(X_train,y_train,epochs=10,validation_split=0.2)
```

```python
y_prob=model.predict(X_test)
```

```python

y_pred=y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
