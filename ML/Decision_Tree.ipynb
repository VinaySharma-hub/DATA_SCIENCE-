# DECISION TREE

Decision tree implementations 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 

```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
iris['data']
```

```python
## It's as Independant feature 
X= pd.DataFrame(iris['data'],columns=['sepal lenth in cm ',
                                   'sepal width', ' petal length','petal width'])
## It's as dependant feauture ]
Y= pd.DataFrame(iris['target'])
X,Y
```

```python
## trainingm the models and train the test split 
from sklearn.model_selection import train_test_split 
X_TRAIN ,X_TEST, Y_TRAIN, Y_TEST = train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
treeclassifier=DecisionTreeClassifier()
```

```python
treeclassifier
```

```python
treeclassifier.fit(X_TRAIN,Y_TRAIN)
```

```python
y_pred=treeclassifier.predict(X_TEST)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(accuracy_score(y_pred,Y_TEST))
print(confusion_matrix(y_pred,Y_TEST))
print(classification_report(Y_TEST,y_pred))
```

```python
### visualising the decision tree fro
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

# Decision Tree classifier Preproning and Hyperparameter Tuning 

```python
from sklearn.tree import DecisionTreeClassifier
param={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2'],
    
               
}
param
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(treeclassifier,param_grid=param,cv=5,scoring='accuracy')
```

```python
grid.fit(X_TRAIN,Y_TRAIN)
```

```python
y_pred=grid.predict(X_TEST)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
print(accuracy_score(y_pred,Y_TEST))
print(confusion_matrix(y_pred,Y_TEST))
print(classification_report(Y_TEST,y_pred))
```

# Diabetes Prediction using decision tree 

```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
df_diabetes=pd.DataFrame(dataset.data,columns=['age','sex','bmi','bp','s1','s2',
's3','s4' ,'s5','s6'])        
df_diabetes                
```

```python
## Independant and dependant features 
X=df_diabetes
Y=dataset['target']
Y
```

```python
# training the datasets fromm scratch by import models and train_test_split
from sklearn.model_selection  import train_test_split
X_TRAIN, X_TEST, Y_TRAIN, Y_TEST=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
X_TRAIN.corr()
```

```python
import seaborn as sns 
plt.Figure(figsize=(15,10))
print(sns.heatmap(X_TRAIN.corr(),annot=True))
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(X_TRAIN,Y_TRAIN)
```

```python
y_pred=regressor.predict(X_TEST)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_pred,Y_TEST))
      
```

## hyperparameter tuning - now we will do it 
### Why we need to do hyperparamteer tuning first of all 
Because model performance heavily depends on these values.

If hyperparameters are not chosen properly:

The model might underfit → too simple → poor accuracy.

Or overfit → too complex → poor generalization on unseen data.

So, hyperparameter tuning = finding the best combination of settings that gives maximum accuracy (or minimum error) on unseen data.

```python
param={
'criterion':['squared_error','friedman_mse','absolute_error'],
'splitter':['best','random'],
'max_depth':[1,2,3,4,5,10,20,25],
'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
regressor=DecisionTreeRegressor()
grid=GridSearchCV(regressor,param_grid=param,cv=5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_TRAIN,Y_TRAIN)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_TEST)
```

```python
print(r2_score(y_pred,Y_TEST))
```
