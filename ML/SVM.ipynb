```python
# 1. svc ( support vector classifier )
# 2. svr ( support vector regression ) 

import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import seaborn as sns
 
```

```python
from sklearn.datasets import make_classification
X,Y = make_classification(n_samples=1000, n_features =2, n_classes=2, n_clusters_per_class=2,n_redundant=0)
```

```python
from sklearn.model_selection import train_test_split
X_TRAIN,X_TEST,Y_TRAIN, Y_TEST =train_test_split(X,Y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel='linear')

svc.fit(X_TRAIN,Y_TRAIN)
```

```python
### Predictions 
y_pred = svc.predict(X_TEST)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(Y_TEST,y_pred
                            ))

print(confusion_matrix(Y_TEST,y_pred))
```

```python
### rbf kernels 
rbf=SVC(kernel='rbf')
```

```python
rbf.fit(X_TRAIN,Y_TRAIN)
y_pred1=rbf.predict(X_TEST)
```

```python
y_pred1
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(Y_TEST,y_pred1
                            ))

print(confusion_matrix(Y_TEST,y_pred1))
```

## Hyperparameter Tuning 

| Parameter                 | Meaning                   | Short Description                                                                                                          |
| ------------------------- | ------------------------- | -------------------------------------------------------------------------------------------------------------------------- |
| **SVC()**                 | Estimator                 | The base model (Support Vector Classifier here) on which hyperparameter tuning will be done.                               |
| **param_grid=param_grid** | Hyperparameter dictionary | Contains the parameters and their possible values to try (e.g., `{C: [1,10], kernel: [linear, rbf]}`).             |
| **refit=True**            | Refit best model          | After finding the best parameters, retrains the model automatically on the entire dataset using the best found parameters. |
| **cv=5**                  | Cross-validation folds    | Number of folds to split data for cross-validation (5-fold CV).                                                            |
| **verbose=3**             | Display level             | Controls output during fitting. Higher number  more detailed logs during training.                                        |

logs- Text output showing the progress of model training and evaluation for each parameter combination.





```python
from sklearn.model_selection import GridSearchCV
## defining parameter range 
param_grid = {'C':[0.1,1,100,1000],
              'gamma':[1,0.1,0.01,0.001,0.0001],
              'kernel':['rbf']}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_TRAIN,Y_TRAIN)
```

```python
grid.best_params_
```

```python
grid.estimator
y_pred2=grid.predict(X_TEST)
```

```python

from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(Y_TEST,y_pred2
                            ))

print(confusion_matrix(Y_TEST,y_pred2))
```
