## K Nearest Neighbour Classifier 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
from sklearn.datasets import make_classification
X,Y= make_classification(
    n_samples=1000 ,## 1000 observations 
    n_features=3, ## total 3 features 
    n_redundant=1 ,# Redundant features are those that don’t add new information — they are linear combinations of the informative features.
    n_classes=2,  ### binary target / label target 
    random_state=999 #random_state is basically a seed value used to control randomness in machine learning functions.
)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_TRAIN , X_TEST, Y_TRAIN, Y_TEST = train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5, algorithm='auto')
```

```python
classifier.fit(X_TRAIN,Y_TRAIN)
```

```python
y_pred=classifier.predict(X_TEST)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(accuracy_score(y_pred,Y_TEST))
print(confusion_matrix(y_pred,Y_TEST))
print(classification_report(y_pred,Y_TEST))
```

# KNearest Regressor 

```python
from sklearn.datasets import make_regression
X1,Y1= make_regression(n_samples=100, n_features=2, noise=10, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
X_TRAIN ,X_TEST ,Y_TRAIN, Y_TEST = train_test_split(X1,Y1,test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor 
regressor=KNeighborsRegressor(n_neighbors=6, algorithm='auto')
regressor.fit(X_TRAIN,Y_TRAIN)
```

```python
y_pred2=regressor.predict(X_TEST)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print(r2_score(Y_TEST,y_pred2))
print(mean_absolute_error(Y_TEST,y_pred2))
print(mean_squared_error(y_pred2,Y_TEST))
```
